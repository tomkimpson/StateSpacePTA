```python
def load_and_plot(f,ax):

    
    #load and plot
    data = np.load(f) 
    x = data[:,0]
    y = data[:,1]



    #sort in order of strain
    p = x.argsort()
    x = x[p]
    y = y[p]

    ax.scatter(x,y)

    #Min detectable strain
    ynew = 10.0 
    xnew = np.interp(ynew, y, x)
    fid = f.split('_')[-1]
    print(f"Minimal detectable strain for beta = {ynew} and f = {fid} is: {xnew}")
```

```python
import matplotlib.pyplot as plt 
import scienceplots
import numpy as np 




plt.style.use('science')
h,w = 12,12
rows = 1
cols = 1
fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(h,w))




for f in list_of_files:
    load_and_plot(f,ax)


ax.set_xscale('log')
ax.set_yscale('log')



fs=20
ax.set_xlabel(r'$h_0$', fontsize=fs)
ax.set_ylabel(r'$\beta$', fontsize=fs)

ax.xaxis.set_tick_params(labelsize=fs-4)
ax.yaxis.set_tick_params(labelsize=fs-4)

ax.axhline(10,linestyle='--', c='0.5')


ax.set_ylim(1e0,2e4)



savefig = 'CanonicalBayesPlot2000_multinoise'
plt.savefig(f"../data/images/{savefig}.png", bbox_inches="tight",dpi=300)

```
