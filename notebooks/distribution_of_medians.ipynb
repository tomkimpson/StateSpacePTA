```python
import json 
import pandas as pd 



def surface_numbers(path):
    # Opening JSON file
    f = open(path)

    # returns JSON object as 
    # a dictionary
    data = json.load(f)

    #Make it a dataframe. Nice for surfacing
    df_posterior = pd.DataFrame(data["posterior"]["content"]) # posterior

    return df_posterior.median()
```

```python
import glob 
list_of_files = sorted(glob.glob("../data/nested_sampling/scp_dir_noise/*.json"))
print(len(list_of_files))
```

```python
import glob 
list_of_files = sorted(glob.glob("../data/nested_sampling/scp_dir_noise/*.json"))



dfs = []


for f in list_of_files:
    print(f)
    df_i = surface_numbers(f)
    dfs.extend([df_i])


df = pd.concat(dfs)
```

```python
import numpy as np 
import scienceplots
plt.style.use('science')


def plot_single_histogram(df,parameter,parameter_label,injection_value,xlims,logscale):

    h,w = 10,10
    rows = 1
    cols = 1
    fig, ax = plt.subplots(nrows=rows, ncols=cols, figsize=(h,w))
    


    df[parameter].hist(ax=ax,bins=20,orientation="vertical",align="left")


    fs=18


    
    


    ax.set_xlabel(parameter_label, fontsize=fs)
    ax.set_ylabel('Counts', fontsize=fs)

    
    ax.xaxis.set_tick_params(labelsize=fs-4)
    ax.yaxis.set_tick_params(labelsize=fs-4)
    ax.grid(False)

    ax.set_xlim(xlims)
   

    ax.axvline(injection_value,c='C2', linestyle='--')

    if logscale:
        ax.set_xscale('log')



    
```

```python
plot_single_histogram(df,'omega_gw',r'$\omega$',5e-7,(1e-9,1e-5),logscale=True)
plot_single_histogram(df,'h',r'$h$',1e-12,(1e-13,1e-11),logscale=True)
```
