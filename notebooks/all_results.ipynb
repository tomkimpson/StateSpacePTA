```python
import json
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import corner
def get_results(path):



    # Opening JSON file
    f = open(path)
    
    # returns JSON object as 
    # a dictionary
    data = json.load(f)


    df_samples = pd.DataFrame(data["samples"]["content"])
    df_posterior = pd.DataFrame(data["posterior"]["content"])

    display(df_samples)
    display(df_posterior)


    median_value = np.median(df_samples[0])
    print("The median value was: ", median_value)



    truths = [5e-7]
    labels = [r"$\omega$"]

    fig = plt.figure(figsize=(10, 10))
    corner.corner(df_samples.to_numpy(),color='C0',show_titles=True,smooth=True, smooth1d=True,labels=labels,truth_color='C1',quantiles=[0.16, 0.84],figsize=(20,12))
    plt.xlim(1e-9,1e-6)
    plt.xscale('log')

##corner.overplot_lines(fig, truths, color="C1")


#plt.savefig("../data/plots/example_corner_plot.png",dpi=300, bbox_inches = "tight")




#fig = corner.corner(y,truths=truths)









```
