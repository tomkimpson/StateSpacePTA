```python
import numpy as np 
num = 2 

seed = 1237
generator = np.random.default_rng(seed)

if num > 1:
    Ω = generator.uniform(low = 1e-7,high=1e-6,size=num)
    Φ0 = generator.uniform(low = 0.0,high=np.pi/2,size=num)
    ψ = generator.uniform(low = 0.0,high=np.pi,size=num)
    ι = generator.uniform(low = 0.0,high=np.pi/2,size=num)
    δ = generator.uniform(low = 0.0,high=np.pi/2,size=num)
    α = generator.uniform(low = 0.0,high=np.pi,size=num)
    h = generator.uniform(low = 5e-15,high=1e-14,size=num)


    injection_parameters = np.array([Ω,Φ0,ψ,ι,δ,α,h]).flatten()

else: 
    injection_parameters = [5e-7,0.20,2.50,1.0,1.0,1.0,5e-15]
```
