## How are our pulsars distributed on the sky?

```python
import pandas as pd
pulsars = pd.read_csv("../data/NANOGrav_pulsars.csv")
print(len(pulsars))    
```

```python
dec = pulsars["DECJD"].to_numpy()
ra = pulsars["RAJD"].to_numpy()
```

```python
print(ra.min())
print(ra.max())
```

```python
import numpy as np 
long1=np.mod((ra+np.pi),2*np.pi)-np.pi #turn RA from 0 - 360 to \pm 180
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")

ax.scatter(long1,dec) #The mollweide projection would require the coordinates in the range -, and -/2., /2.


BH_dec = 0.40
BH_ra = 4.80 
long1_BH=np.mod((BH_ra+np.pi),2*np.pi)-np.pi

ax.scatter(long1_BH,BH_dec,c='r')
ax.grid(True)
```
